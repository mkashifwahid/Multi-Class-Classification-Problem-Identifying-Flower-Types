##### Step 2 - We will start off by importing all of the classes and functions we will need:

```python
# Import Libararies

import numpy
import numpy as np
import pandas
from keras.models import Sequential, Model
from keras import layers
from keras.layers import Dense, Input
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn import model_selection
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from keras import regularizers
from keras.utils import np_utils
import tensorflow as tf
```

##### Extended of Step 2 - Next, we can initialize the random number generator to ensure that we always get the same results when executing this code. This will help if we are debugging:

```python
# fix random seed for reproducibility

seed = 7
numpy.random.seed(seed)
```

##### Extended of Step 2 - Now we can load the dataset using pandas and split the columns into 60 input variables (X) and 1 output variable (Y). We use pandas to load the data because it easily handles strings (the output variable), whereas attempting to load the data directly using NumPy would be more difficult

```python
# load Sonar dataset using Pandas
dataframe = pandas.read_csv("iris.csv", header=None)
dataset = dataframe.values
X = dataset[:,0:4].astype(float)
Y = dataset[:,4]
```

```python
# Shape of Data

dataset.shape
```

```python
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
# convert integers to dummy variables (i.e. one hot encoded)
dummy_y = np_utils.to_categorical(encoded_Y)
```

##### Step 3 - Defining the function that creates our baseline model

```python
# baseline model
def baseline_model():
        # create model, write code below
        model = Sequential()
        model.add(Dense(8, activation="relu", input_shape=(4,)))
        model.add(Dense(3, activation="softmax"))
          
        # Compile model, write code below
        model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
        return model
```

##### Step 4. Evaluate The Model with k-Fold Cross Validation

```python
estimator = KerasClassifier(build_fn=baseline_model, epochs=200, batch_size=5, verbose=0)
kfold = model_selection.KFold(n_splits=10, shuffle=True, random_state=seed)
results = cross_val_score(estimator, X, dummy_y, cv=kfold)
print("Baseline: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

##### Step 8. Rewriting the code using the Keras Functional API

```python
# Main model By Keras api
def create_model_by_function_api():
    # create model, write code below
    inputs = Input(shape=(4,))
    x_input = Dense(8, activation='relu')(inputs)
    x_input = Dense(3, activation='relu')(x_input)
    predictions = Dense(3, activation = 'softmax')(x_input)
    model = Model(inputs=inputs, outputs=predictions)
        
    # Compile model, write code below
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
numpy.random.seed(seed)
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasClassifier(build_fn=create_larger, epochs=26, batch_size=5, verbose=0)))
pipeline = Pipeline(estimators)
kfold = model_selection.KFold(n_splits=10, shuffle=True, random_state=seed)
results = cross_val_score(pipeline, X, dummy_y, cv=kfold)
print("Function Api Model : %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

##### Step 9. Rewriting the code by doing Model Subclassing

```python

class SubClass_Model(tf.keras.Model):
    def __init__(self):
        super(SubClass_Model, self).__init__()
        self.dense1 = tf.keras.layers.Dense(8, activation='relu')
        self.dense2 = tf.keras.layers.Dense(3, activation='relu')
        self.dense3 = tf.keras.layers.Dense(3, activation='softmax')

    def call(self, inputs):
        x_input = self.dense1(inputs)
        x_input = self.dense2(x_input)
        outputs = self.dense3(x_input)
        return outputs
```

```python
# Sub Class model By Keras api
def create_model_by_SubClass():
    # create model, write code below
    model = SubClass_Model()
        
    # Compile model, write code below
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
numpy.random.seed(seed)
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasClassifier(build_fn=create_model_by_SubClass, epochs=26, batch_size=5, verbose=0)))
pipeline = Pipeline(estimators)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
results = cross_val_score(pipeline, X, encoded_Y, cv=kfold)
print("Sub Class Api Model : %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

```python
history_dict = validation_scores.history
validation_score = np.average(history_dict['acc'])
#validation_score = np.average(validation_scores)
print("Model using K-Fold without skitlearn : %.2f%% " % (validation_score*100))
```
